```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle

```

```python
df= pd.read_csv('E:\Music Recommendation System\model.ipynb')
```

```python
df['combined'] = (
    df['song_name'].astype(str) + " " +
    df['artist'].astype(str) + " " +
    df['album'].astype(str) + " " +
    df['genre'].astype(str) 
)
```

```python
df['combined']
```

```python
tfid = TfidfVectorizer(stop_words = 'english')
vectors = tfid.fit_transform(df['combined'])
```

```python
vectors 
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(song):
    song = song.lower()

    try:
        index = df[df['song_name'].str.lower()== song.index[0]]
    except:
        return []
    distance = similarity[index]
    song_list = sorted(list(enumerate(distance)),
                        reverse = True,key = lambda x:x[1])[1:6]
    recommendation=[]
    for i in song_list:
        recommendation.append(df.iloc[i[0]]["song_name"])
    return recommendation                        
```

```python
pickle.dump(df,open("song_df.pkl",'wb'))
pickle.dump(similarity,open("similarity.pkl",'wb'))
pickle.dump(tfid,open("tfid.pkl",'wb'))
```
